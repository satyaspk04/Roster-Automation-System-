```python
import pandas as pd
import numpy as np
import math
from google.colab import files

# Upload the file
uploaded = files.upload()

# Get the file name (assuming you uploaded the correct file)
file_name = list(uploaded.keys())[0]

# Read the CSV file into a DataFrame
emp = pd.read_csv(file_name)

# Print the columns of the DataFrame
print(emp.columns)

# Input parameters
LM = int(input('Enter Total Number of employees: '))  # total headcount
HC = math.ceil(LM * 5 / 7)  # required headcount
WO = LM - HC  # employees in layoff

print("Enter Required employees: ", HC)
print("Number of employees under layoff are: ", WO)

vol = int(input('Enter shipment volume: '))  # number of shipments
tph = int(input('Enter shipments per hour: '))  # number of shipments cleared in an hour
working_hrs = vol / tph  # total number of hours to clear the shipments
print("working hours:", int(working_hrs))

DS = int(input("Enter day shift Head Count(%): "))
day_shift_employees = math.ceil(HC * DS / 100)
print("Employees required in dayshift: ", day_shift_employees)

NS = int(input('Enter night shift Head Count(%):'))
night_shift_employees = math.ceil(HC * NS / 100)
print("Employeees required in night shift: ", night_shift_employees)

# Define the days of the week
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Function to generate a schedule with 5 workdays, 2 consecutive days off, and a shift assignment
def generate_schedule(shift):
    # Possible consecutive days off
    days_off_options = [
        ('Monday', 'Tuesday'), ('Tuesday', 'Wednesday'), ('Wednesday', 'Thursday'),
        ('Thursday', 'Friday'), ('Friday', 'Saturday'), ('Saturday', 'Sunday'),
        ('Sunday', 'Monday')
    ]
    # Select a random pair of consecutive days off
    days_off = days_off_options[np.random.randint(0, len(days_off_options))]

    # Create a work schedule with shift assignment
    work_schedule = {day: shift for day in days}
    work_schedule[days_off[0]] = 'No Work'
    work_schedule[days_off[1]] = 'No Work'

    return work_schedule

# Initialize the DataFrame to store the schedule
schedule_columns = [f'{day} (Week 1)' for day in days] + [f'{day} (Week 2)' for day in days]
schedule = pd.DataFrame(index=[f'employee-{i}' for i in range(1, LM + 1)], columns=schedule_columns)

# Assign the schedule to each employee
for i in range(1, day_shift_employees + 1):
    employee_schedule = generate_schedule('Morning')
    for day in days:
        schedule.at[f'employee-{i}', f'{day} (Week 1)'] = employee_schedule[day]
        schedule.at[f'employee-{i}', f'{day} (Week 2)'] = employee_schedule[day]

for i in range(day_shift_employees + 1, day_shift_employees + night_shift_employees + 1):
    employee_schedule = generate_schedule('Night')
    for day in days:
        schedule.at[f'employee-{i}', f'{day} (Week 1)'] = employee_schedule[day]
        schedule.at[f'employee-{i}', f'{day} (Week 2)'] = employee_schedule[day]

# Mark layoff employees with a single "Week_off" entry and leave the rest empty
for i in range(HC + 1, LM + 1):
    schedule.at[f'employee-{i}', f'{days[0]} (Week 1)'] = 'Week_off'
    for day in days[1:]:
        schedule.at[f'employee-{i}', f'{day} (Week 1)'] = ''
    for day in days:
        schedule.at[f'employee-{i}', f'{day} (Week 2)'] = ''

# Combine the working schedule and layoff information
flattened_schedule = schedule.reset_index().rename(columns={'index': 'Employee'})

# Print the full schedule
print(flattened_schedule.head())

# Save the full schedule to a CSV file
flattened_schedule.to_csv('full_employee_schedule.csv', index=False)

# Download the file to the local machine in Google Colab
files.download('full_employee_schedule.csv')
```
